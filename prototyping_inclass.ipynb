```python
# Transform table into Pandas DataFrame
import pandas as pd

#load table from local CSV file with old data
df = pd.read_csv('progressboard.csv')
df = df.drop(450)
df = df[df.user != 'dominikb1888']
df

# Create PivotTable with the correctly aggreagted data
```

```python
dff = df[['user','session','status']]


# How do we generate a new column in our data from function applied on two other columns

dff.loc[(dff['status'] == 'completed') | (dff['status'] == 'rework') , 'status'] = 1
dff.loc[(dff['status'] == 'failing') | (dff['status'] == 'not-started'), 'status'] = 0
dff
```

```python
import numpy as np

pt = pd.pivot_table(dff, values='status', index=['user'], columns=['session'], aggfunc=np.sum, fill_value=0)
pt
```

```python
import matplotlib.pyplot as plt
import matplotlib_inline
import seaborn as sns

matplotlib_inline.backend_inline.set_matplotlib_formats("svg")

# Feed DataFrame to Seaborn Heatmap
sns.set(rc = {'figure.figsize':(6,15)})
sns.heatmap(pt, cmap="YlGnBu", linewidths=3, annot=True)

# normalize the color gradient on each session
# instead of feeding absolute data, 
# we need to feed relative data and 
# annote with absolute values

```

```python
dfo = pd.read_csv('progressboard.csv')
```

```python
dfo = dfo[dfo['user'] == 'dominikb1888']
dfo = dfo.drop(450)
dfo = dfo[['session','exercise']].drop_duplicates(subset=None, keep='first', inplace=False)
dfo
```

```python
# group data by session and count exercise
mapping = dfo.groupby(['session']).count().to_dict()['exercise']
mapping
```

```python
pt_rel = pd.pivot_table(dff, values='status', index=['user'], columns=['session'], aggfunc=np.sum, fill_value=0)

for (name, abs_value) in pt_rel.iteritems():
    pt_rel[name] = abs_value / mapping[name]

pt_rel

```

```python
import matplotlib.pyplot as plt
import matplotlib_inline
import seaborn as sns

matplotlib_inline.backend_inline.set_matplotlib_formats("svg")

# Feed DataFrame to Seaborn Heatmap
sns.set(rc = {'figure.figsize':(6,15)})
sns.heatmap(pt_rel, cmap="YlGnBu", linewidths=3, annot = pt)
plt.savefig('heatmap.png')
```
